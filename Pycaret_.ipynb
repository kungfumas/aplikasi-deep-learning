```python
import pandas as pd
from pycaret.classification import *
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rashida048/Datasets/master/Heart.csv')
```

```python
df.head()
```

```python
cat_features = ['Sex', 'ChestPain', 'Fbs', 'RestECG', 'ExAng', 'Thal']
```

```python
experiment = setup(df, target='AHD', categorical_features=cat_features)
```

```python
best_model = compare_models()
```

```python
df.tail()
```

```python
predict_model(best_model, df.tail())
```

```python
save_model(best_model, model_name='logistic-model')
```

```python
model = load_model('logistic-model')
```

```python
model.predict(df.tail())
```
