##For 40X

```
base_dir = '/content/drive/My Drive/BreaKHis_v1/tf_files/40X/'
#image_path=glob('/content/drive/My Drive/Datasets/ICIAR2018_BACH_Challenge/train/Train_set/**/*.png', recursive=True)
classone=glob(base_dir+'benign/*.png' ,recursive=True)
classtwo=glob(base_dir+'malignant/*.png' ,recursive=True)
```

```
labels='Benign','Malignant'
sizes=[len(classone),len(classtwo)]
colors=['lightskyblue','red']
plt.pie(sizes,labels=labels, colors=colors, autopct='%1.1f%%' ,shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

```
len(classone)
```

```
len(classtwo)
```

```
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 6), subplot_kw=dict(aspect="equal"))

recipe = ["626 images Benign",
          "1370 images Malignant"]

data = [float(x.split()[0]) for x in recipe]
classes = [x.split()[-1] for x in recipe]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} Samples)".format(pct, absolute)

wedges, texts, autotexts = ax.pie(sizes, autopct=lambda pct: func(pct, data),textprops=dict(color="w"),startangle=140)
ax.legend(wedges, classes,
          title="Class Name",
          loc="center left",
          prop={'size': 14},
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=12, weight="bold")
ax.set_title('40X Data Distribution',fontsize=20)
plt.show()
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
labels = ['Benign', 'Malignant']
sizes=[len(classone),len(classtwo)]
width = 0.35
ax.bar(labels,sizes,width,color='b')
ax.set_ylabel('No of images')
ax.set_xlabel('No of Classes')
ax.set_title('40X Classes Data Distribution')
plt.show()
```
